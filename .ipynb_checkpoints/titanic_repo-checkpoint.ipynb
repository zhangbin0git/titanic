## 1. 有哪些因素会让船上的人生还率更高？

数据集中的因素包括：
1	Survived	生存	0 =否，1 =是
2	Pclass	票类	1 = 1，2 = 2，3 = 3
3	Sex	性别	
4	Age	年龄	
5	SibSp	泰坦尼克号上的兄弟姐妹/配偶	
6	Parch	泰坦尼克号上的父母/孩子的数量	
7	Ticket	票号	
8	Fare	乘客票价	
9	Cabin	客舱号码	
10	Embarked	登船港口	C =瑟堡，Q =皇后镇，S =南安普敦

### 我分析可能的因素为1、性别  2、年龄   3、票类/票价   4.孩子有父母和女性有配偶或兄弟生还率将增大


```python
# 导入相关包数据
%pylab inline
import numpy as py
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
# 导入数据集
data = pd.read_csv('titanic-data.csv')
data['Fare_1'] = ''
data['Age_1'] = ''
```

```python
# 显示前三条数据
data.head(50)
```

```python
# 总人数
pepole_sum = data.count()['Name']
pepole_sum 
```

```python
def fare_cmp(para):
    if para > 200:
        return 'A'
    elif para > 150:
        return 'B'
    elif para > 100:
        return 'C'
    elif para >50:
        return 'D'
    else:
        return 'E'
```

```python
def age_cmp(para):
    if para > 18:
        return 'A'
    elif para > 10:
        return 'B'
    else:
        return 'C'
```

```python
data['Fare_1'] = data['Fare'].map(fare_cmp)
```

```python
data['Age_1']=data['Age'].map(age_cmp)
```

```python
data
```

```python
def ratio(data_cache, key_col, live):
    # 返回key_col分类的live的比率，live为1为生存率，0为死,cata为具体类别
    result = []
    group_cache = data_cache.groupby([key_col, 'Survived'])
    cata_key = data_cache.groupby([key_col, 'Survived']).size().index.levels[0].values
    for i in cata_key:
        print '分类基数' + str(i) + ':' + str(group_cache.size()[i].sum())
        result.append(float(group_cache.size()[i, live]) / group_cache.size()[i].sum())
    return pd.Series(result, index=cata_key).plot(kind='bar')
```

```python
# 分类中SEX类别生存率
ratio(data, 'Sex', 1)
```

```python
# 分类中Fare类别生存率
ratio(data, 'Fare_1', 1)
```

```python
# 分类中Age类别生存率
ratio(data, 'Age_1', 1) 
```

```python
# 分类中Pclass类别生存率
ratio(data, 'Pclass', 1) 
```

```python
# 分类中Embarked类别生存率
ratio(data, 'Embarked', 1) 
```

```python
def ratio_three(data_cache, key_1, key_2, key_3, live):
    # 返回key_col分类的live的比率，live为1为生存率，0为死,cata为具体类别
    result = []
    group_cache = data_cache.groupby([key_1, key_2, key_3, 'Survived'])
    cata_key = data_cache.groupby([key_1, key_2, key_3, 'Survived']).size().index.levels[0].values
    for i in cata_key:
        print '分类基数' + str(i) + ':' + str(group_cache.size()[i].sum())
        result.append(float(group_cache.size()[i, live]) / group_cache.size()[i].sum())
    return pd.Series(result, index=cata_key).plot(kind='bar')
```
